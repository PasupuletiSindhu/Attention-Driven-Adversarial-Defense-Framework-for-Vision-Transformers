```python
from __future__ import annotations

import math
import os
import random
import time
from dataclasses import dataclass
from pathlib import Path
from typing import Dict, List, Optional, Tuple

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import timm
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, Subset
from torchvision import datasets, transforms
from torchvision.models import resnet50, ResNet50_Weights
from torchvision.utils import make_grid
from torchattacks import FGSM, PGD

import transformers 

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_theme(style='darkgrid')

FORCE_DEVICE = 'cuda'  # Setting to 'cuda' for Kaggle GPU, 'mps' for Apple Silicon, or 'cpu' for CPU

if FORCE_DEVICE:
    device = torch.device(FORCE_DEVICE)
    print(f" Forced device: {device}")
elif torch.cuda.is_available():
    device = torch.device('cuda')
    print(f" Using CUDA GPU: {torch.cuda.get_device_name(0)}")
elif hasattr(torch.backends, 'mps') and torch.backends.mps.is_available():
    device = torch.device('mps')
    print(f" Using MPS (Apple Silicon GPU)")
else:
    device = torch.device('cpu')
    print(f"  Using CPU (no GPU available)")

print(f"Using torch {torch.__version__} on device: {device}")

#Configuring Classes and Functions

@dataclass
class ExperimentConfig:
    data_root: Path = Path('data/cifar10')
    image_size: int = 224
    batch_size: int = 64
    num_workers: int = min(8, os.cpu_count() or 2)
    seed: int = 42
    device: str = str(device)
    epsilon_fgsm: float = 8.0 / 255.0
    epsilon_pgd: float = 8.0 / 255.0
    pgd_alpha: float = 2.0 / 255.0
    pgd_steps: int = 10
    adv_train_steps: int = 10
    adv_train_alpha: float = 2.0 / 255.0
    adv_train_epsilon: float = 8.0 / 255.0
    adv_train_epochs: int = 5

def set_global_seed(seed: int) -> None:
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

cfg = ExperimentConfig()
set_global_seed(cfg.seed)
print(f"Experiment configuration: {cfg}")

# Loading Data and creating Data Loaders
imagenet_mean = (0.485, 0.456, 0.406)
imagenet_std = (0.229, 0.224, 0.225)

basic_transform = transforms.Compose([
    transforms.Resize(cfg.image_size),
    transforms.ToTensor(),
    transforms.Normalize(imagenet_mean, imagenet_std),
])

raw_train = datasets.CIFAR10(
    root=cfg.data_root,
    train=True,
    download=True,
    transform=basic_transform,
)
raw_test = datasets.CIFAR10(
    root=cfg.data_root,
    train=False,
    download=True,
    transform=basic_transform,
)

class_names = raw_train.classes
print(f"Loaded CIFAR-10 train/test splits with sizes: {len(raw_train)} / {len(raw_test)}")

train_loader = DataLoader(
    raw_train,
    batch_size=cfg.batch_size,
    shuffle=True,
    num_workers=cfg.num_workers,
    pin_memory=True,
)

test_loader = DataLoader(
    raw_test,
    batch_size=cfg.batch_size,
    shuffle=False,
    num_workers=cfg.num_workers,
    pin_memory=True,
)

print('Dataset sizes -> train/test:', len(raw_train), len(raw_test))

# Loading model's functions

def load_vit(device: str) -> nn.Module:
    model = timm.create_model('vit_base_patch16_224', pretrained=True)
    in_features = model.head.in_features
    model.head = nn.Linear(in_features, 10)
    return model.to(device)

def load_resnet(device: str) -> nn.Module:
    weights = ResNet50_Weights.IMAGENET1K_V2
    model = resnet50(weights=weights)
    model.fc = nn.Linear(model.fc.in_features, 10)
    return model.to(device)

def count_params(model: nn.Module) -> float:
    return sum(p.numel() for p in model.parameters()) / 1e6

# Fine tuning Model

def fine_tune_model(model: nn.Module, train_loader: DataLoader, test_loader: DataLoader, epochs: int = 3, lr: float = 1e-4, device: str = 'cpu', 
                    model_name: str = "Model", val_batches: int = 5) -> nn.Module:
    # Quickly fine-tuning on CIFAR-10 with progress tracking
    device_obj = torch.device(device) if isinstance(device, str) else device
    model = model.to(device_obj)
    model.train()
    optimizer = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=0.01)
    criterion = nn.CrossEntropyLoss()
    
    num_batches = len(train_loader)
    print(f"Fine-tuning {model_name} for {epochs} epochs ({num_batches} batches/epoch)...")
    if str(device_obj) == 'cpu':
        print("  Running on CPU - this will be slow")
    
    total_start = time.time()
    for epoch in range(epochs):
        epoch_start = time.time()
        total_loss = 0.0
        correct = 0
        total = 0
        
        for batch_idx, (images, labels) in enumerate(train_loader):
            images, labels = images.to(device_obj), labels.to(device_obj)
            optimizer.zero_grad()
            logits = model(images)
            loss = criterion(logits, labels)
            loss.backward()
            optimizer.step()
            
            total_loss += loss.item()
            preds = logits.argmax(dim=1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)
            
            if (batch_idx + 1) % 10 == 0 or (batch_idx + 1) == num_batches:
                elapsed = time.time() - epoch_start
                rate = (batch_idx + 1) / elapsed if elapsed > 0 else 0
                remaining = (num_batches - batch_idx - 1) / rate if rate > 0 else 0
                print(f"  Batch {batch_idx + 1}/{num_batches} | Loss: {loss.item():.4f} | "
                      f"ETA: {remaining/60:.1f} min")
        
        train_acc = correct / total
        avg_loss = total_loss / num_batches
        
        model.eval()
        val_correct = 0
        val_total = 0
        with torch.no_grad():
            for batch_idx, (images, labels) in enumerate(test_loader):
                if batch_idx >= val_batches:
                    break
                images, labels = images.to(device_obj), labels.to(device_obj)
                logits = model(images)
                preds = logits.argmax(dim=1)
                val_correct += (preds == labels).sum().item()
                val_total += labels.size(0)
        val_acc = val_correct / val_total
        model.train()
        
        epoch_time = time.time() - epoch_start
        print(f"Epoch {epoch+1}/{epochs} ({epoch_time/60:.1f} min) - "
              f"Loss: {avg_loss:.4f}, Train Acc: {train_acc:.4f}, Val Acc: {val_acc:.4f}")
        if epoch < epochs - 1:
            remaining_epochs = epochs - epoch - 1
            est_remaining = (epoch_time * remaining_epochs) / 60
            print(f"  Estimated time remaining: {est_remaining:.1f} minutes")
    
    total_time = time.time() - total_start
    print(f" {model_name} fine-tuning complete in {total_time/60:.1f} minutes\n")
    model.eval()
    return model

# Evaluating Functions
@torch.no_grad()
def evaluate(model: nn.Module, loader: DataLoader, device: str, max_batches: Optional[int] = None, 
             model_name: str = "Model") -> Dict[str, float]:
    # Evaluating model on clean data
    model.eval()
    top1 = 0
    n = 0
    
    num_batches = len(loader) if max_batches is None else min(max_batches, len(loader))
    print(f"Evaluating {model_name} on {num_batches} batches...")
    
    start_time = time.time()
    for batch_idx, (images, labels) in enumerate(loader):
        if max_batches is not None and batch_idx >= max_batches:
            break
            
        images, labels = images.to(device), labels.to(device)
        logits = model(images)
        preds = logits.argmax(dim=1)
        top1 += (preds == labels).sum().item()
        n += labels.size(0)
        
        if (batch_idx + 1) % 5 == 0 or (batch_idx + 1) == num_batches:
            elapsed = time.time() - start_time
            rate = (batch_idx + 1) / elapsed if elapsed > 0 else 0
            remaining = (num_batches - batch_idx - 1) / rate if rate > 0 else 0
            current_acc = top1 / n
            print(f"  Batch {batch_idx + 1}/{num_batches} | Acc: {current_acc:.4f} | "
                  f"ETA: {remaining/60:.1f} min")
    
    acc = top1 / n
    total_time = time.time() - start_time
    print(f"{model_name} evaluation complete: {acc:.4f} ({top1}/{n}) in {total_time:.1f}s\n")
    return {'top1': acc}

def build_attack(model: nn.Module, attack_type: str):
    attack_type = attack_type.lower()
    if attack_type == 'fgsm':
        return FGSM(model, eps=cfg.epsilon_fgsm)
    if attack_type == 'pgd':
        return PGD(model, eps=cfg.epsilon_pgd, alpha=cfg.pgd_alpha, steps=cfg.pgd_steps, random_start=True)
    raise ValueError(f"Unsupported attack {attack_type}")

def evaluate_under_attack(model: nn.Module, loader: DataLoader, attack_type: str, 
                          max_batches: Optional[int] = None, 
                          sub_batch_size: int = 16,
                          use_cpu_for_attacks: bool = False) -> float:
    # Evaluating model under adversarial attack with memory-efficient processing
    if use_cpu_for_attacks:
        attack_device = torch.device('cpu')
        eval_device = torch.device('cpu')
        model = model.cpu()
        model.eval()
        attack = build_attack(model, attack_type)
    else:
        attack_device = torch.device(cfg.device) if isinstance(cfg.device, str) else cfg.device
        eval_device = torch.device(cfg.device) if isinstance(cfg.device, str) else cfg.device
        model = model.to(eval_device)
        model.eval()
        attack = build_attack(model, attack_type)
    
    total = 0
    correct = 0
    
    num_batches = len(loader) if max_batches is None else min(max_batches, len(loader))
    device_str = "CPU" if use_cpu_for_attacks else str(attack_device)
    print(f"Evaluating {attack_type.upper()} attack on {num_batches} batches (sub-batch: {sub_batch_size}, device: {device_str})...")
    
    for batch_idx, (images, labels) in enumerate(loader):
        if max_batches is not None and batch_idx >= max_batches:
            break
        
        batch_size = images.size(0)
        batch_correct = 0
        
        for i in range(0, batch_size, sub_batch_size):
            end_idx = min(i + sub_batch_size, batch_size)
            sub_images = images[i:end_idx]
            sub_labels = labels[i:end_idx]
            
            sub_images_attack = sub_images.to(attack_device)
            sub_labels_attack = sub_labels.to(attack_device)
            adv_sub_images = attack(sub_images_attack.clone(), sub_labels_attack)
            
            adv_sub_images_eval = adv_sub_images.to(eval_device)
            sub_labels_eval = sub_labels.to(eval_device)
            
            with torch.no_grad():
                logits = model(adv_sub_images_eval)
                preds = logits.argmax(dim=1)
                batch_correct += (preds == sub_labels_eval).sum().item()
            
            del adv_sub_images, adv_sub_images_eval, logits, preds
            del sub_images_attack, sub_labels_attack, sub_labels_eval
            if attack_device.type == 'mps':
                torch.mps.empty_cache()
            elif attack_device.type == 'cuda':
                torch.cuda.empty_cache()
            if eval_device.type == 'mps':
                torch.mps.empty_cache()
            elif eval_device.type == 'cuda':
                torch.cuda.empty_cache()
        
        correct += batch_correct
        total += batch_size
        
        if (batch_idx + 1) % 5 == 0:
            print(f"  Batch {batch_idx + 1}/{num_batches}, Current Acc: {correct/total:.4f}")
    
    acc = correct / total
    print(f"  {attack_type.upper()} accuracy: {acc:.4f} ({correct}/{total})")
    return acc

# Attention Recorder Class
class AttentionRecorder:
    def __init__(self, model: nn.Module):
        self.model = model
        self.handles: List[torch.utils.hooks.RemovableHandle] = []
        self.activations: List[torch.Tensor] = []

    def _hook_fn(self, module, input, output):
        # For timm ViT, we need to capture attention from the forward pass
        # The attention weights are computed in the attention module
        # We'll hook into the attention module's forward method
        pass

    def _attention_hook(self, module, input, output):
        # Hook that captures attention weights after softmax
        # In timm ViT, attn_drop receives the attention weights after softmax
        if output is not None:
            # output shape: (batch, heads, seq_len, seq_len) for attention weights
            self.activations.append(output.detach().cpu())

    def __enter__(self):
        # Hook into each attention block's attention mechanism
        for block in self.model.blocks:
            try:
                handle = block.attn.attn_drop.register_forward_hook(self._attention_hook)
                self.handles.append(handle)
            except AttributeError:
                # If attn_drop doesn't work, then hook into the attention module itself
                try:
                    # Hook into the forward method of the attention module
                    original_forward = block.attn.forward
                    
                    def hook_forward(x):
                        B, N, C = x.shape
                        qkv = block.attn.qkv(x).reshape(B, N, 3, block.attn.num_heads, C // block.attn.num_heads).permute(2, 0, 3, 1, 4)
                        q, k, v = qkv[0], qkv[1], qkv[2]
                        attn = (q @ k.transpose(-2, -1)) * block.attn.scale
                        attn = attn.softmax(dim=-1)
                        self.activations.append(attn.detach().cpu())
                        return original_forward(x)
                    
                    block.attn.forward = hook_forward
                    self.handles.append(None)  # Mark that we modified forward
                except:
                    pass
        return self

    def __exit__(self, exc_type, exc_val, exc_tb):
        # Restoring original forward methods if we modified them
        for i, handle in enumerate(self.handles):
            if handle is not None:
                handle.remove()
            else:
                # Restoring original forward (we need to track which blocks we modified)
                pass
        self.handles.clear()
        self.activations.clear()

@torch.no_grad()
def get_vit_attentions(model: nn.Module, images: torch.Tensor, sample_index: int = 0) -> torch.Tensor:
    # Extracting attention maps from ViT model. Returns: (layers, heads, seq_len, seq_len) tensor
    model.eval()
    
    # Storing attention activations
    activations = []
    original_forwards = []
    
    # Creating hooked forward methods for each attention block
    # Using a factory function to properly capture each block's attention module
    def create_attention_hook(attn_mod, activations_list):
        # Factory function to create a hook that properly captures the attention module
        def hooked_forward(x, attn_mask=None):
            B, N, C = x.shape
            qkv = attn_mod.qkv(x).reshape(B, N, 3, attn_mod.num_heads, C // attn_mod.num_heads).permute(2, 0, 3, 1, 4)
            q, k, v = qkv[0], qkv[1], qkv[2]
            attn = (q @ k.transpose(-2, -1)) * attn_mod.scale
            if attn_mask is not None:
                attn = attn + attn_mask
            attn = attn.softmax(dim=-1)
            activations_list.append(attn.detach().cpu())
            attn = attn_mod.attn_drop(attn)
            x = (attn @ v).transpose(1, 2).reshape(B, N, C)
            x = attn_mod.proj(x)
            x = attn_mod.proj_drop(x)
            return x
        return hooked_forward
    
    for block_idx, block in enumerate(model.blocks):
        original_forward = block.attn.forward
        attn_module = block.attn
        hooked_forward = create_attention_hook(attn_module, activations)
        original_forwards.append((block_idx, original_forward))
        block.attn.forward = hooked_forward
    
    try:
        _ = model(images)
        
        if len(activations) == 0:
            raise ValueError("No attention activations captured. Check hook placement.")
        attn = torch.stack(activations)
        attn = attn[:, sample_index]
        
    finally:
        for block_idx, original_forward in original_forwards:
            model.blocks[block_idx].attn.forward = original_forward
    
    return attn

# Attention Metrics Functions
def normalize_attention(attn: torch.Tensor, eps: float = 1e-8) -> torch.Tensor:
    attn = attn / (attn.sum(dim=-1, keepdim=True) + eps)
    return attn

def compute_kl_divergence(clean_attn: torch.Tensor, adv_attn: torch.Tensor, eps: float = 1e-8) -> torch.Tensor:
    clean = normalize_attention(clean_attn, eps)
    adv = normalize_attention(adv_attn, eps)
    kl = (clean * ((clean + eps).log() - (adv + eps).log())).sum(dim=-1)
    return kl.mean(dim=-1)

def compute_attention_entropy(attn: torch.Tensor, eps: float = 1e-8) -> torch.Tensor:
    attn = normalize_attention(attn, eps)
    entropy = -(attn * (attn + eps).log()).sum(dim=-1)
    return entropy.mean(dim=-1)

def compute_head_variance(attn: torch.Tensor, eps: float = 1e-8) -> torch.Tensor:
    attn = normalize_attention(attn, eps)
    head_energy = attn.mean(dim=-1)
    return head_energy.var(dim=-1)

def summarize_attention_drift(clean_attn: torch.Tensor, adv_attn: torch.Tensor) -> pd.DataFrame:
    kl = compute_kl_divergence(clean_attn, adv_attn)
    entropy_clean = compute_attention_entropy(clean_attn)
    entropy_adv = compute_attention_entropy(adv_attn)
    var_clean = compute_head_variance(clean_attn)
    var_adv = compute_head_variance(adv_attn)

    layers, heads = kl.shape
    data = {
        'layer': np.repeat(np.arange(layers), heads),
        'head': np.tile(np.arange(heads), layers),
        'KL': kl.cpu().numpy().reshape(-1),
        'entropy_clean': entropy_clean.cpu().numpy().reshape(-1),
        'entropy_adv': entropy_adv.cpu().numpy().reshape(-1),
        'variance': (var_adv - var_clean).cpu().numpy().reshape(-1),
    }
    return pd.DataFrame(data)

# Visualizing Functions
def denormalize_image(tensor, mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)):
    # Denormalizing image tensor for visualization
    mean = torch.tensor(mean).view(3, 1, 1)
    std = torch.tensor(std).view(3, 1, 1)
    return tensor * std + mean

def visualize_adversarial_examples(clean_images, adv_images_fgsm, adv_images_pgd, 
                                   labels, class_names, num_images=4):
    # Visualizing clean vs adversarial examples side by side
    num_images = min(num_images, clean_images.size(0))
    
    fig, axes = plt.subplots(num_images, 4, figsize=(16, 4 * num_images))
    if num_images == 1:
        axes = axes.reshape(1, -1)
    
    fig.suptitle('Clean vs Adversarial Examples Comparison', fontsize=16, y=0.995)
    
    for i in range(num_images):
        clean_img = denormalize_image(clean_images[i].cpu()).clamp(0, 1)
        fgsm_vit_img = denormalize_image(adv_images_fgsm['vit'][i].cpu()).clamp(0, 1)
        pgd_vit_img = denormalize_image(adv_images_pgd['vit'][i].cpu()).clamp(0, 1)
        fgsm_resnet_img = denormalize_image(adv_images_fgsm['resnet'][i].cpu()).clamp(0, 1)
        
        with torch.no_grad():
            vit_model.eval()
            resnet_model.eval()
            
            clean_logits_vit = vit_model(clean_images[i:i+1].to(cfg.device))
            fgsm_logits_vit = vit_model(adv_images_fgsm['vit'][i:i+1].to(cfg.device))
            pgd_logits_vit = vit_model(adv_images_pgd['vit'][i:i+1].to(cfg.device))
            clean_logits_resnet = resnet_model(clean_images[i:i+1].to(cfg.device))
            fgsm_logits_resnet = resnet_model(adv_images_fgsm['resnet'][i:i+1].to(cfg.device))
            
            clean_pred_vit = clean_logits_vit.argmax(dim=1).item()
            fgsm_pred_vit = fgsm_logits_vit.argmax(dim=1).item()
            pgd_pred_vit = pgd_logits_vit.argmax(dim=1).item()
            clean_pred_resnet = clean_logits_resnet.argmax(dim=1).item()
            fgsm_pred_resnet = fgsm_logits_resnet.argmax(dim=1).item()
        
        true_label = labels[i].item()
        true_class = class_names[true_label]
        
        axes[i, 0].imshow(clean_img.permute(1, 2, 0).numpy())
        axes[i, 0].set_title(f'Clean\nTrue: {true_class}\nViT: {class_names[clean_pred_vit]}\nResNet: {class_names[clean_pred_resnet]}', 
                            fontsize=10)
        axes[i, 0].axis('off')
        
        axes[i, 1].imshow(fgsm_vit_img.permute(1, 2, 0).numpy())
        axes[i, 1].set_title(f'FGSM (ViT)\nPred: {class_names[fgsm_pred_vit]}', 
                            fontsize=10, color='red' if fgsm_pred_vit != true_label else 'green')
        axes[i, 1].axis('off')
        
        axes[i, 2].imshow(pgd_vit_img.permute(1, 2, 0).numpy())
        axes[i, 2].set_title(f'PGD (ViT)\nPred: {class_names[pgd_pred_vit]}', 
                            fontsize=10, color='red' if pgd_pred_vit != true_label else 'green')
        axes[i, 2].axis('off')
        
        axes[i, 3].imshow(fgsm_resnet_img.permute(1, 2, 0).numpy())
        axes[i, 3].set_title(f'FGSM (ResNet)\nPred: {class_names[fgsm_pred_resnet]}', 
                            fontsize=10, color='red' if fgsm_pred_resnet != true_label else 'green')
        axes[i, 3].axis('off')
    
    plt.tight_layout()
    
    figures_dir = Path('figures')
    figures_dir.mkdir(parents=True, exist_ok=True)
    plt.savefig(figures_dir / 'adversarial_examples_comparison.png', dpi=200, bbox_inches='tight')
    print(f"Saved visualization to {figures_dir / 'adversarial_examples_comparison.png'}")
    
    plt.show()

def plot_cls_attention(clean_attn: torch.Tensor, adv_attn: torch.Tensor, attack_name: str, save_path: Path) -> None:
    SELECTED_HEADS = [0, 3, 7]
    LAYER_TO_VISUALIZE = -1
    
    fig, axes = plt.subplots(len(SELECTED_HEADS), 2, figsize=(10, 2.2 * len(SELECTED_HEADS)))
    fig.suptitle(f'CLS Attention Drift ({attack_name})', fontsize=14)

    for row, head_idx in enumerate(SELECTED_HEADS):
        clean_row = clean_attn[LAYER_TO_VISUALIZE, head_idx, 0].cpu().numpy().reshape(1, -1)
        adv_row = adv_attn[LAYER_TO_VISUALIZE, head_idx, 0].cpu().numpy().reshape(1, -1)

        axes[row, 0].imshow(clean_row, aspect='auto', cmap='magma')
        axes[row, 0].set_title(f'Head {head_idx} - Clean')
        axes[row, 0].set_ylabel('CLS')
        axes[row, 0].set_xticks([])
        axes[row, 0].set_yticks([])

        im = axes[row, 1].imshow(adv_row, aspect='auto', cmap='magma')
        axes[row, 1].set_title(f'Head {head_idx} - {attack_name}')
        axes[row, 1].set_xticks([])
        axes[row, 1].set_yticks([])

    fig.colorbar(im, ax=axes, fraction=0.02, pad=0.02)
    plt.tight_layout()
    plt.savefig(save_path, dpi=200, bbox_inches='tight')
    plt.show()

# Main Execution Coding starts
# Loading  models
print("LOADING MODELS")
vit_model = load_vit(cfg.device)
resnet_model = load_resnet(cfg.device)
print(f"Loaded ViT-B/16 with {count_params(vit_model):.2f}M params")
print(f"Loaded ResNet-50 with {count_params(resnet_model):.2f}M params")

# Fine-tuning configuration
SKIP_FINE_TUNING = False
MAX_BATCHES_PER_EPOCH = None
FINE_TUNE_EPOCHS = 1
USE_CPU_FOR_TRAINING = False

if SKIP_FINE_TUNING:
    print("SKIPPING FINE-TUNING")
else:
    if MAX_BATCHES_PER_EPOCH:
        subset_indices = list(range(0, min(MAX_BATCHES_PER_EPOCH * cfg.batch_size, len(raw_train))))
        train_subset = Subset(raw_train, subset_indices)
        fast_train_loader = DataLoader(train_subset, batch_size=cfg.batch_size, shuffle=True, 
                                       num_workers=cfg.num_workers, pin_memory=True)

        print("FAST FINE-TUNING (Subset Mode)")
        print(f"Using {len(train_subset)} samples, {MAX_BATCHES_PER_EPOCH} batches/epoch")
        print(f"Epochs: {FINE_TUNE_EPOCHS}\n")
    else:
        fast_train_loader = train_loader
        print("FULL FINE-TUNING")
    
    if USE_CPU_FOR_TRAINING:
        training_device = 'cpu'
        print("Using CPU for training\n")
    else:
        training_device = cfg.device
        print(f"Using {training_device} for training\n")
    
    print("Fine-tuning ViT...")
    vit_model = vit_model.to(torch.device(training_device) if isinstance(training_device, str) else training_device)
    vit_model = fine_tune_model(vit_model, fast_train_loader, test_loader, 
                               epochs=FINE_TUNE_EPOCHS, lr=1e-4, device=training_device,
                               model_name="ViT", val_batches=5)
    print("ViT Fine-tuning complete")
    
    print("\nFine-tuning ResNet...")
    resnet_model = resnet_model.to(torch.device(training_device) if isinstance(training_device, str) else training_device)
    resnet_model = fine_tune_model(resnet_model, fast_train_loader, test_loader, 
                                   epochs=FINE_TUNE_EPOCHS, lr=1e-4, device=training_device,
                                   model_name="ResNet", val_batches=5)
    print("ResNet Fine-tuning complete")

# Baseline evaluation
print("BASELINE EVALUATION ON CLEAN DATA")
EVAL_MAX_BATCHES = 10
SKIP_RESNET_EVAL = False

vit_clean_metrics = evaluate(vit_model, test_loader, cfg.device, 
                            max_batches=EVAL_MAX_BATCHES, model_name="ViT")

if SKIP_RESNET_EVAL:
    resnet_clean_metrics = {'top1': 0.0}
else:
    resnet_clean_metrics = evaluate(resnet_model, test_loader, cfg.device, 
                                   max_batches=EVAL_MAX_BATCHES, model_name="ResNet")

print("SUMMARY")
print(f'ViT clean accuracy:   {vit_clean_metrics["top1"]:.4f}')
if not SKIP_RESNET_EVAL:
    print(f'ResNet clean accuracy: {resnet_clean_metrics["top1"]:.4f}')
# Generating adversarial examples

print("GENERATING ADVERSARIAL EXAMPLES")
NUM_SAMPLE_IMAGES = 4
USE_CPU_FOR_SAMPLES = False

sample_images, sample_labels = next(iter(test_loader))
sample_images = sample_images[:NUM_SAMPLE_IMAGES]
sample_labels = sample_labels[:NUM_SAMPLE_IMAGES]

if USE_CPU_FOR_SAMPLES:
    sample_device = torch.device('cpu')
    clean_images = sample_images.to(sample_device)
    sample_labels = sample_labels.to(sample_device)
    vit_model_cpu = vit_model.cpu()
    resnet_model_cpu = resnet_model.cpu()
    
    adv_images_fgsm = {
        'vit': build_attack(vit_model_cpu, 'fgsm')(clean_images.clone(), sample_labels).detach(),
        'resnet': build_attack(resnet_model_cpu, 'fgsm')(clean_images.clone(), sample_labels).detach(),
    }
    adv_images_pgd = {
        'vit': build_attack(vit_model_cpu, 'pgd')(clean_images.clone(), sample_labels).detach(),
        'resnet': build_attack(resnet_model_cpu, 'pgd')(clean_images.clone(), sample_labels).detach(),
    }
    
    vit_model = vit_model.to(cfg.device)
    resnet_model = resnet_model.to(cfg.device)
    clean_images = clean_images.to(cfg.device)
    sample_labels = sample_labels.to(cfg.device)
    adv_images_fgsm = {k: v.to(cfg.device) for k, v in adv_images_fgsm.items()}
    adv_images_pgd = {k: v.to(cfg.device) for k, v in adv_images_pgd.items()}
else:
    clean_images = sample_images.to(cfg.device)
    sample_labels = sample_labels.to(cfg.device)
    
    adv_images_fgsm = {
        'vit': build_attack(vit_model, 'fgsm')(clean_images.clone(), sample_labels).detach(),
        'resnet': build_attack(resnet_model, 'fgsm')(clean_images.clone(), sample_labels).detach(),
    }
    adv_images_pgd = {
        'vit': build_attack(vit_model, 'pgd')(clean_images.clone(), sample_labels).detach(),
        'resnet': build_attack(resnet_model, 'pgd')(clean_images.clone(), sample_labels).detach(),
    }

print(f'Generated FGSM and PGD adversarial batches for ViT & ResNet ({NUM_SAMPLE_IMAGES} images)')

# Ensuring models are on correct device
device_obj = torch.device(cfg.device) if isinstance(cfg.device, str) else cfg.device
vit_model = vit_model.to(device_obj)
resnet_model = resnet_model.to(device_obj)
clean_images = clean_images.to(device_obj)
adv_images_fgsm = {k: v.to(device_obj) for k, v in adv_images_fgsm.items()}
adv_images_pgd = {k: v.to(device_obj) for k, v in adv_images_pgd.items()}

# Visualizing adversarial examples
print("VISUALIZING ADVERSARIAL EXAMPLES")
visualize_adversarial_examples(clean_images, adv_images_fgsm, adv_images_pgd, 
                               sample_labels, class_names, num_images=NUM_SAMPLE_IMAGES)

# Evaluating under attack
print("EVALUATING ADVERSARIAL ROBUSTNESS")
MAX_EVAL_BATCHES = 30  # Increased from 5 to 30 for more reliable metrics (~1920 samples)
SUB_BATCH_SIZE = 16
USE_CPU_FOR_ATTACKS = False

vit_fgsm_acc = evaluate_under_attack(vit_model, test_loader, 'fgsm', max_batches=MAX_EVAL_BATCHES, sub_batch_size=SUB_BATCH_SIZE, use_cpu_for_attacks=USE_CPU_FOR_ATTACKS)
vit_pgd_acc = evaluate_under_attack(vit_model, test_loader, 'pgd', max_batches=MAX_EVAL_BATCHES, sub_batch_size=SUB_BATCH_SIZE, use_cpu_for_attacks=USE_CPU_FOR_ATTACKS)
resnet_fgsm_acc = evaluate_under_attack(resnet_model, test_loader, 'fgsm', max_batches=MAX_EVAL_BATCHES, sub_batch_size=SUB_BATCH_SIZE, use_cpu_for_attacks=USE_CPU_FOR_ATTACKS)
resnet_pgd_acc = evaluate_under_attack(resnet_model, test_loader, 'pgd', max_batches=MAX_EVAL_BATCHES, sub_batch_size=SUB_BATCH_SIZE, use_cpu_for_attacks=USE_CPU_FOR_ATTACKS)

print('Adversarial Robustness Summary')
print(f'ViT   - FGSM: {vit_fgsm_acc:.4f}, PGD: {vit_pgd_acc:.4f}')
print(f'ResNet - FGSM: {resnet_fgsm_acc:.4f}, PGD: {resnet_pgd_acc:.4f}')

# Extracting attention maps
print("EXTRACTING ATTENTION MAPS")
clean_attn_maps = get_vit_attentions(vit_model, clean_images)
fgsm_attn_maps = get_vit_attentions(vit_model, adv_images_fgsm['vit'])
pgd_attn_maps = get_vit_attentions(vit_model, adv_images_pgd['vit'])
print('Attention tensor shape:', clean_attn_maps.shape)

# Computing attention metrics
print("COMPUTING ATTENTION METRICS")
fgsm_metrics_df = summarize_attention_drift(clean_attn_maps, fgsm_attn_maps)
pgd_metrics_df = summarize_attention_drift(clean_attn_maps, pgd_attn_maps)
attention_drift_df = pd.concat([
    fgsm_metrics_df.assign(attack='FGSM'),
    pgd_metrics_df.assign(attack='PGD'),
], ignore_index=True)
print('Attention drift summary shape:', attention_drift_df.shape)

# Visualizing attention
print("VISUALIZING ATTENTION")
figures_dir = Path('figures')
figures_dir.mkdir(parents=True, exist_ok=True)
plot_cls_attention(clean_attn_maps, fgsm_attn_maps, 'FGSM', figures_dir / 'cls_attention_fgsm.png')
plot_cls_attention(clean_attn_maps, pgd_attn_maps, 'PGD', figures_dir / 'cls_attention_pgd.png')

# Adversarial training
print("ADVERSARIAL TRAINING")
# Configuration for adversarial training batches
ADV_TRAIN_MAX_BATCHES = 150

vit_model_adv = load_vit(cfg.device)
optimizer = torch.optim.AdamW(vit_model_adv.parameters(), lr=1e-4, weight_decay=0.01)
criterion = nn.CrossEntropyLoss()
pgd_attack = PGD(vit_model_adv, eps=cfg.adv_train_epsilon, alpha=cfg.adv_train_alpha, steps=cfg.adv_train_steps)

if ADV_TRAIN_MAX_BATCHES:
    print(f"Starting PGD adversarial training for {cfg.adv_train_epochs} epochs...")
    print(f"Using {ADV_TRAIN_MAX_BATCHES} batches/epoch (~{ADV_TRAIN_MAX_BATCHES * cfg.batch_size} samples/epoch)\n")
else:
    print(f"Starting PGD adversarial training for {cfg.adv_train_epochs} epochs...")
    print(f"Using full dataset (~{len(train_loader)} batches/epoch)\n")

for epoch in range(cfg.adv_train_epochs):
    vit_model_adv.train()
    total_loss = 0.0
    correct = 0
    total = 0
    num_batches_processed = 0

    for batch_idx, (images, labels) in enumerate(train_loader):
        if ADV_TRAIN_MAX_BATCHES and batch_idx >= ADV_TRAIN_MAX_BATCHES:
            break
        images = images.to(cfg.device)
        labels = labels.to(cfg.device)

        adv_images = pgd_attack(images.clone(), labels)
        optimizer.zero_grad()
        logits = vit_model_adv(adv_images)
        loss = criterion(logits, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        preds = logits.argmax(dim=1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

        if (batch_idx + 1) % 100 == 0:
            print(f"Epoch {epoch+1}/{cfg.adv_train_epochs}, Batch {batch_idx+1}, Loss: {loss.item():.4f}, Acc: {100*correct/total:.2f}%")

    avg_loss = total_loss / len(train_loader)
    train_acc = correct / total
    print(f"Epoch {epoch+1} completed - Avg Loss: {avg_loss:.4f}, Train Acc: {train_acc:.4f}")

checkpoint_path = Path('vit_pgd_trained.pth')
torch.save(vit_model_adv.state_dict(), checkpoint_path)
print(f"Saved adversarially trained model to {checkpoint_path}")

# Recomputing metrics after training
print("RECOMPUTING METRICS AFTER TRAINING")
if checkpoint_path.exists():
    vit_model_adv.load_state_dict(torch.load(checkpoint_path, map_location=cfg.device))
    print("Loaded adversarially trained ViT model")
else:
    print("Warning: Checkpoint not found. Using current model state.")

vit_model_adv.eval()

print("Generating new adversarial examples using the adversarially trained model...")
fgsm_attack_adv = build_attack(vit_model_adv, 'fgsm')
pgd_attack_adv = build_attack(vit_model_adv, 'pgd')

adv_images_fgsm_after = fgsm_attack_adv(clean_images.clone(), sample_labels).detach()
adv_images_pgd_after = pgd_attack_adv(clean_images.clone(), sample_labels).detach()

# Extracting attention from adversarially trained model on new adversarial examples
clean_attn_after = get_vit_attentions(vit_model_adv, clean_images)
fgsm_attn_after = get_vit_attentions(vit_model_adv, adv_images_fgsm_after)
pgd_attn_after = get_vit_attentions(vit_model_adv, adv_images_pgd_after)

fgsm_metrics_after = summarize_attention_drift(clean_attn_after, fgsm_attn_after)
pgd_metrics_after = summarize_attention_drift(clean_attn_after, pgd_attn_after)

print("Before training - FGSM KL mean:", fgsm_metrics_df['KL'].mean())
print("After training - FGSM KL mean:", fgsm_metrics_after['KL'].mean())
print("Before training - PGD KL mean:", pgd_metrics_df['KL'].mean())
print("After training - PGD KL mean:", pgd_metrics_after['KL'].mean())

# Plotting before/after comparison
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
for idx, (attack_name, before_df, after_df) in enumerate([
    ('FGSM', fgsm_metrics_df, fgsm_metrics_after),
    ('PGD', pgd_metrics_df, pgd_metrics_after),
]):
    before_kl_by_layer = before_df.groupby('layer')['KL'].mean()
    after_kl_by_layer = after_df.groupby('layer')['KL'].mean()
    axes[idx].plot(before_kl_by_layer.index, before_kl_by_layer.values, 'o-', label='Before Training', linewidth=2)
    axes[idx].plot(after_kl_by_layer.index, after_kl_by_layer.values, 's-', label='After Training', linewidth=2)
    axes[idx].set_xlabel('Layer', fontsize=12)
    axes[idx].set_ylabel('Mean KL Divergence', fontsize=12)
    axes[idx].set_title(f'{attack_name} Attention Drift: Before vs After Training', fontsize=13)
    axes[idx].legend()
    axes[idx].grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(figures_dir / 'attention_drift_before_after.png', dpi=200, bbox_inches='tight')
plt.show()

# Comprehensive attention metrics visualization
print("COMPREHENSIVE ATTENTION METRICS VISUALIZATION")

# 1. KL Divergence Distribution Plots
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Attention Metrics: Before vs After Adversarial Training', fontsize=16, y=0.995)

# KL Divergence distributions
for idx, (attack_name, before_df, after_df) in enumerate([('FGSM', fgsm_metrics_df, fgsm_metrics_after),('PGD', pgd_metrics_df, pgd_metrics_after),]):
    axes[0, idx].hist(before_df['KL'].values, bins=30, alpha=0.6, label='Before Training', color='blue', edgecolor='black')
    axes[0, idx].hist(after_df['KL'].values, bins=30, alpha=0.6, label='After Training', color='red', edgecolor='black')
    axes[0, idx].set_xlabel('KL Divergence', fontsize=12)
    axes[0, idx].set_ylabel('Frequency', fontsize=12)
    axes[0, idx].set_title(f'{attack_name} - KL Divergence Distribution', fontsize=13)
    axes[0, idx].legend()
    axes[0, idx].grid(True, alpha=0.3)
    
    # Entropy comparison
    axes[1, idx].scatter(before_df['entropy_clean'].values, before_df['entropy_adv'].values, 
                         alpha=0.5, label='Before Training', s=20, color='blue')
    axes[1, idx].scatter(after_df['entropy_clean'].values, after_df['entropy_adv'].values, 
                         alpha=0.5, label='After Training', s=20, color='red', marker='s')
    axes[1, idx].plot([0, 10], [0, 10], 'k--', alpha=0.3, linewidth=1)
    axes[1, idx].set_xlabel('Clean Entropy', fontsize=12)
    axes[1, idx].set_ylabel('Adversarial Entropy', fontsize=12)
    axes[1, idx].set_title(f'{attack_name} - Attention Entropy Comparison', fontsize=13)
    axes[1, idx].legend()
    axes[1, idx].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(figures_dir / 'attention_metrics_comprehensive.png', dpi=200, bbox_inches='tight')
plt.show()

# 2. Entropy and Variance Summary Plots
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Attention Entropy and Variance: Before vs After Training', fontsize=16, y=0.995)

for idx, (attack_name, before_df, after_df) in enumerate([('FGSM', fgsm_metrics_df, fgsm_metrics_after),('PGD', pgd_metrics_df, pgd_metrics_after),]):
    before_entropy_by_layer = before_df.groupby('layer')['entropy_adv'].mean()
    after_entropy_by_layer = after_df.groupby('layer')['entropy_adv'].mean()
    axes[0, idx].plot(before_entropy_by_layer.index, before_entropy_by_layer.values, 
                     'o-', label='Before Training', linewidth=2, markersize=8)
    axes[0, idx].plot(after_entropy_by_layer.index, after_entropy_by_layer.values, 
                     's-', label='After Training', linewidth=2, markersize=8)
    axes[0, idx].set_xlabel('Layer', fontsize=12)
    axes[0, idx].set_ylabel('Mean Adversarial Entropy', fontsize=12)
    axes[0, idx].set_title(f'{attack_name} - Attention Entropy by Layer', fontsize=13)
    axes[0, idx].legend()
    axes[0, idx].grid(True, alpha=0.3)

    before_var_by_layer = before_df.groupby('layer')['variance'].mean()
    after_var_by_layer = after_df.groupby('layer')['variance'].mean()
    axes[1, idx].plot(before_var_by_layer.index, before_var_by_layer.values, 
                     'o-', label='Before Training', linewidth=2, markersize=8)
    axes[1, idx].plot(after_var_by_layer.index, after_var_by_layer.values, 
                     's-', label='After Training', linewidth=2, markersize=8)
    axes[1, idx].set_xlabel('Layer', fontsize=12)
    axes[1, idx].set_ylabel('Mean Head Variance', fontsize=12)
    axes[1, idx].set_title(f'{attack_name} - Head Variance by Layer', fontsize=13)
    axes[1, idx].legend()
    axes[1, idx].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(figures_dir / 'attention_entropy_variance.png', dpi=200, bbox_inches='tight')
plt.show()

# 3. Side-by-side attention heatmap comparison
print("\nGenerating side-by-side attention comparison...")
num_layers_to_show = 4 
num_heads_to_show = 4

fig, axes = plt.subplots(num_layers_to_show, num_heads_to_show * 2, figsize=(20, 8))
fig.suptitle('Attention Maps: Before vs After Adversarial Training (FGSM Attack)', fontsize=16, y=0.995)

for layer_idx in range(num_layers_to_show):
    for head_idx in range(num_heads_to_show):
        # Before training - attention maps are [layers, heads, tokens, tokens], no batch dimension
        attn_before = fgsm_attn_maps[layer_idx, head_idx, 1:, 1:].cpu().numpy()  # Exclude CLS token
        im1 = axes[layer_idx, head_idx * 2].imshow(attn_before, cmap='hot', aspect='auto')
        axes[layer_idx, head_idx * 2].set_title(f'L{layer_idx}H{head_idx} Before', fontsize=9)
        axes[layer_idx, head_idx * 2].axis('off')
        
        # After training
        attn_after = fgsm_attn_after[layer_idx, head_idx, 1:, 1:].cpu().numpy()
        im2 = axes[layer_idx, head_idx * 2 + 1].imshow(attn_after, cmap='hot', aspect='auto')
        axes[layer_idx, head_idx * 2 + 1].set_title(f'L{layer_idx}H{head_idx} After', fontsize=9)
        axes[layer_idx, head_idx * 2 + 1].axis('off')

plt.tight_layout()
plt.savefig(figures_dir / 'attention_side_by_side_comparison.png', dpi=200, bbox_inches='tight')
plt.show()

# Printing summary statistics
print("ATTENTION METRICS SUMMARY")
print("\nBefore Training:")
print(f"  FGSM - Mean KL: {fgsm_metrics_df['KL'].mean():.4f}, Mean Entropy (Adv): {fgsm_metrics_df['entropy_adv'].mean():.4f}, Mean Variance: {fgsm_metrics_df['variance'].mean():.4f}")
print(f"  PGD  - Mean KL: {pgd_metrics_df['KL'].mean():.4f}, Mean Entropy (Adv): {pgd_metrics_df['entropy_adv'].mean():.4f}, Mean Variance: {pgd_metrics_df['variance'].mean():.4f}")

print("\nAfter Training:")
print(f"  FGSM - Mean KL: {fgsm_metrics_after['KL'].mean():.4f}, Mean Entropy (Adv): {fgsm_metrics_after['entropy_adv'].mean():.4f}, Mean Variance: {fgsm_metrics_after['variance'].mean():.4f}")
print(f"  PGD  - Mean KL: {pgd_metrics_after['KL'].mean():.4f}, Mean Entropy (Adv): {pgd_metrics_after['entropy_adv'].mean():.4f}, Mean Variance: {pgd_metrics_after['variance'].mean():.4f}")

print("FINAL ROBUSTNESS COMPARISON")
vit_clean_acc = evaluate(vit_model, test_loader, cfg.device)['top1']
vit_adv_clean_acc = evaluate(vit_model_adv, test_loader, cfg.device)['top1']
vit_fgsm_acc = evaluate_under_attack(vit_model, test_loader, 'fgsm', max_batches=MAX_EVAL_BATCHES, 
                                     sub_batch_size=SUB_BATCH_SIZE, use_cpu_for_attacks=USE_CPU_FOR_ATTACKS)
vit_adv_fgsm_acc = evaluate_under_attack(vit_model_adv, test_loader, 'fgsm', max_batches=MAX_EVAL_BATCHES, 
                                          sub_batch_size=SUB_BATCH_SIZE, use_cpu_for_attacks=USE_CPU_FOR_ATTACKS)
vit_pgd_acc = evaluate_under_attack(vit_model, test_loader, 'pgd', max_batches=MAX_EVAL_BATCHES, 
                                    sub_batch_size=SUB_BATCH_SIZE, use_cpu_for_attacks=USE_CPU_FOR_ATTACKS)
vit_adv_pgd_acc = evaluate_under_attack(vit_model_adv, test_loader, 'pgd', max_batches=MAX_EVAL_BATCHES, 
                                        sub_batch_size=SUB_BATCH_SIZE, use_cpu_for_attacks=USE_CPU_FOR_ATTACKS)
resnet_clean_acc = evaluate(resnet_model, test_loader, cfg.device, max_batches=MAX_EVAL_BATCHES)['top1']
resnet_fgsm_acc = evaluate_under_attack(resnet_model, test_loader, 'fgsm', max_batches=MAX_EVAL_BATCHES, 
                                        sub_batch_size=SUB_BATCH_SIZE, use_cpu_for_attacks=USE_CPU_FOR_ATTACKS)
resnet_pgd_acc = evaluate_under_attack(resnet_model, test_loader, 'pgd', max_batches=MAX_EVAL_BATCHES, 
                                       sub_batch_size=SUB_BATCH_SIZE, use_cpu_for_attacks=USE_CPU_FOR_ATTACKS)

comparison_data = {
    'Model': ['ViT (Pretrained)', 'ViT (Adv Trained)', 'ResNet-50'],
    'Clean Accuracy': [f'{vit_clean_acc:.4f}', f'{vit_adv_clean_acc:.4f}', f'{resnet_clean_acc:.4f}'],
    'FGSM Robustness': [f'{vit_fgsm_acc:.4f}', f'{vit_adv_fgsm_acc:.4f}', f'{resnet_fgsm_acc:.4f}'],
    'PGD Robustness': [f'{vit_pgd_acc:.4f}', f'{vit_adv_pgd_acc:.4f}', f'{resnet_pgd_acc:.4f}'],
    'FGSM Drop': [
        f'{(vit_clean_acc - vit_fgsm_acc)*100:.2f}%',
        f'{(vit_adv_clean_acc - vit_adv_fgsm_acc)*100:.2f}%',
        f'{(resnet_clean_acc - resnet_fgsm_acc)*100:.2f}%',
    ],
    'PGD Drop': [
        f'{(vit_clean_acc - vit_pgd_acc)*100:.2f}%',
        f'{(vit_adv_clean_acc - vit_adv_pgd_acc)*100:.2f}%',
        f'{(resnet_clean_acc - resnet_pgd_acc)*100:.2f}%',
    ],
}

comparison_df = pd.DataFrame(comparison_data)
print("ROBUSTNESS COMPARISON: ViT vs ResNet-50")
print(comparison_df.to_string(index=False))

vit_avg_kl_before = attention_drift_df.groupby('attack')['KL'].mean()
print("\nViT Attention Drift (Before Training):")
print(f"  FGSM Mean KL: {vit_avg_kl_before.get('FGSM', 0):.4f}")
print(f"  PGD Mean KL: {vit_avg_kl_before.get('PGD', 0):.4f}")

vit_avg_kl_after = pd.concat([
    fgsm_metrics_after.assign(attack='FGSM'),
    pgd_metrics_after.assign(attack='PGD'),
]).groupby('attack')['KL'].mean()
print("\nViT Attention Drift (After Training):")
print(f"  FGSM Mean KL: {vit_avg_kl_after.get('FGSM', 0):.4f}")
print(f"  PGD Mean KL: {vit_avg_kl_after.get('PGD', 0):.4f}")

print("ALL EXPERIMENTS COMPLETE")
```
